# Graph analysis

```python
import networkx as nx 

def draw_graph(graph):
    """
    Function that draws the graph, using networkx package
    """
    G=nx.Graph()
    G.add_edges_from(graph.edges())
    nx.draw(G, with_labels = True, node_size=600, node_color='#00b4d9')

def defin_graph(set_edges):
    """
    Function that defines the graph as we need it.
    """
    g=set1.Graph()
    for i in set_edges:
        g.addEdge(i[0],i[1])
    return (g)
```

```python
def right(q):
    """
    This function checks if q is a path.
    """
    if q==None:
        return False
    for i in range(1,len(q)):
        if q[i][0]!=q[i-1][1]:
            return False
    return True
```

```python
def test(s0,start):
    """
    This function performs the algorithms and print both the result and the right() check.
    """
    
    g0=defin_graph(s0)
    
    draw_graph(g0)
    
    print("Fleury - Recursive")
    f=g0.fleury_rec(start)
    print(f)
    print(right(f))
    
    g0=defin_graph(s0)
    
    print("Hierholzer - Recursive")
    unvisited0=g0.edges()
    h=g0.hierholzer2_rec(start,unvisited0,index=0,final=[])
    print(h)
    print(right(h))
    
    g0=defin_graph(s0)
    
    print("Fleury - Iterative")
    l=len(g0.single_edges())
    r=g0.fleury_it(start,l)
    print(r)
    print(right(r))
    
    g0=defin_graph(s0)
    
    print("Hierholzer - Iterative")
    g0=defin_graph(s0)
    h=g0.hierholzer2_it(start)
    print(h)
    print(right(h))
```

If you want to check that our algorithm works also on random graph and starting from any vertex there is the following cell, in which we generate a random graph using the netwrokx package and perform our algorithms if the graph is well.  
Be aware, you can change the parameters but it may take time because the probability of getting a valid graph are very low. So the more the edges, the more the repetitions needed.

```python
from networkx.generators.random_graphs import erdos_renyi_graph

count=0               # number of graphs checked
for a in range(0,50):  # repeat 10 times
    n = 10            # Number of vertices
    p = 0.7          # Probability of creating an edge
    g = erdos_renyi_graph(n, p)

    new=set1.Graph()
    for i in g.edges:
        v1,v2=tuple(i)
        new.addEdge(v1,v2)
    ed=new.edges()
    
    if new.ep_check()==False:
        continue
    
    print("Fleury - Recursive")
    for i in new.vertices():
        new=defin_graph(ed)
        print(right(new.fleury_rec(i)))
    
    print("Hierholzer - Recursive")
    for i in new.vertices():
        new=defin_graph(ed)
        unvisited=new.edges()
        print(right(new.hierholzer2_rec(i,unvisited,index=0,final=[])))
    
    print("Fleury - Iterative")
    for i in new.vertices():
        new=defin_graph(ed)
        l=len(new.single_edges())
        print(right(new.fleury_it(i,l)))
    
    print("Hierholzer - Iterative")
    for i in new.vertices():
        new=defin_graph(ed)
        print(right(new.hierholzer2_it(i)))
    count+=1
print(count)
```

# Time analysis

```python
def util(ed,start,alg):
    
    if alg=="fleuryrec":
        g=defin_graph(ed)
        g.fleury_rec(start)
        
    elif alg=="hierholzerrec":
        g=defin_graph(ed)
        unvisited=g.edges()
        g.hierholzer2_rec(start,unvisited,final=[])
        
    elif alg=="fleuryit":
        g=defin_graph(ed)
        l=len(g.single_edges())
        g.fleury_it(start,l)
        
    elif alg=="hierholzerit":
        g=defin_graph(ed)
        g.hierholzer2_it(start)
        
    elif alg=="hierholzerdeq":
        g=defin_graph_deq(ed)
        g.hierholzer2_deq(start)
    else:
        raise AttributeError("Algorithm not found")
```

```python
def time_analysis1(ed,start):
    g=defin_graph(ed)
    
    print("Fleury - Recursive")
    %timeit g.fleury_rec(start)
    
    #g=defin_graph(ed)
    
    print("Hierholzer - Recursive")
    unvisited=g.edges()
    %timeit g.hierholzer2_rec(start,unvisited,final=[])
    
    print("Fleury - Iterative")
    l=len(g.single_edges())
    %timeit g.fleury_it(start,l)
    
    print("Hierholzer - Iterative")
    %timeit g.hierholzer2_it(start)
    
```

```python
def time_analysis2(ed,start):
    g=defin_graph(ed)
    
    print("Fleury - Recursive")
    %timeit util(ed,start,"fleuryrec")
    
    #g=defin_graph(ed)
    
    print("Hierholzer - Recursive")
    
    %timeit util(ed,start,"hierholzerrec")
    
    print("Fleury - Iterative")
    %timeit util(ed,start,"fleuryit")
    
    print("Hierholzer - Iterative")
    %timeit util(ed,start,"hierholzerit")
    
    print("Hierholzer - Deque")
    %timeit util(ed,start,"hierholzerdeq")
```
